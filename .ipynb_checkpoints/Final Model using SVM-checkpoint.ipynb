```python
## Here we will check for the total word count, character count, get the total number of stopwords and perform visual analysis
```

```python
def _get_wordcounts(x):
    length = len(str(x).split())
    return length
```

```python
def _get_charcounts(x):
    s = x.split()
    x = ''.join(s)
    return len(x)
```

```python
df['char_counts'] = df['Text'].apply(lambda x: _get_charcounts(x))
df['word_counts'] = df['Text'].apply(lambda x: _get_wordcounts(x))
```

```python
df
```

```python
def _get_stopwords_counts(x):
    l = len([t for t in x.split() if t in stopwords])
    return l
```

```python
def _get_avg_wordlength(x):
    count = _get_charcounts(x)/_get_wordcounts(x)
    return count
```

```python
df['avg_wordlength'] = df['Text'].apply(lambda x: _get_avg_wordlength(x))
```

```python
def _get_value_counts(df, col):
    text = ' '.join(df[col])
    text = text.split()
    freq = pd.Series(text).value_counts()
    return freq
```

```python
def get_clean(x):
    x = str(x).lower().replace('\\', ' ').replace('_', ' ').replace('.', ' ')
    x = kgp.cont_exp(x)
    x = kgp.remove_emails(x)
    x = kgp.remove_urls(x)
    x = kgp.remove_rt(x)
    x = kgp.remove_accented_chars(x)
    x = kgp.remove_special_chars(x)
    x = kgp.remove_dups_char(x)
    return x
```

```python
text = df['textstop']
y = df['Category']
```

```python
tfidf = TfidfVectorizer()
X = text
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
X_train.shape, X_test.shape
```

```python
##Creation of a pipeline
```

```python
pipe = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf',LinearSVC() )
])
```

```python
##Hyperparameter Tuning
```

```python
hyperparameters = {
    'tfidf__max_df': (0.5, 1.0),
    'tfidf__ngram_range': ((1,1), (1,2)),
    'tfidf__use_idf': (True, False),
    'tfidf__analyzer': ('word', 'char', 'char_wb')
}
```

```python
clf = GridSearchCV(pipe, hyperparameters, n_jobs=-1, cv = None)
```

```python

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```
